#### Let's create a sample vector array

```python
class KNN:
    def __init__(self, dataset:pd.DataFrame, k:int):
        self.iris_dataset = dataset
        self.k = k
        self.model = None
        self.output = None

    def vectorizer(self):
        vec_x = self.iris_dataset[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].to_numpy()
        vec_y = self.iris_dataset[['species']].to_numpy()

        self.model = vec_x
        self.output = vec_y

    def cosine_sim(self, vec):
        dot_product = np.dot(self.model, vec)
        magnitude_vector = np.linalg.norm(vec)
        magnitude_matrix_rows = np.linalg.norm(self.model, axis=1)
        sims = dot_product / (magnitude_matrix_rows * magnitude_vector)
        return sims

    def predict(self, vector: np.ndarray):
        cos_sim = self.cosine_sim(vector)
        
```

```python
knn = KNN(iris_dataset.sample(150), 3)
```
